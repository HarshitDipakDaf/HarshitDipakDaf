```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/91914/Downloads/Data_companies/companies.csv")
```

```python
df.head()
```

```python
#Step 1: Checking the percentage of NaN (null) values present in each feature
nan_percentage = df.isnull().mean() * 100
print("Percentage of NaN values in each feature:\n", nan_percentage)
```

```python
# Step 2: Removing duplicates
df = df.drop_duplicates()
```

```python
# Step 3: Dropping columns which have NaN values
threshold = 50  
columns_to_drop = nan_percentage[nan_percentage > threshold].index
df = df.drop(columns=columns_to_drop)
```

```python
# Step 4: Remove unnecessary and corrupted data
# Assuming unnecessary columns are identified
unnecessary_columns = ['normalized_name', 'entity_id', 'short_description']
df = df.drop(columns=unnecessary_columns, errors='ignore')  # 'errors' set to ignore in case columns are already dropped
```

```python
# Step 5: Data Labelling 
# Labeling target variable 'status' to numerical values
status_mapping = {'Operating': 0, 'IPO': 1, 'Acquired': 2, 'Closed': 3}
df['status'] = df['status'].map(status_mapping)
```
